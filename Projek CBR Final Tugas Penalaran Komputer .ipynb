```python
gejala_dipilih = []
for i in range(len(gejala)):
    jawab = input("Apakah Mengalami "+gejala[i]+"(Y/T)?")
    if jawab.upper()=="Y":
        gejala_dipilih.append(True)
    else:
        gejala_dipilih.append(False)
print("Gejala Dipilih",gejala_dipilih)
jml_gejala_dipilih = 0
jml_gejala_kasus = 0
jml_gejala_cocok = 0
nilai_hasil = []
penyakit_kasus = []
for i in range (len(kasus)):
    jml_gejala_dipilih = 0
    jml_gejala_kasus = 0
    jml_gejala_cocok = 0
    penyakit_kasus.append("")
    for j in range(len(gejala)):
        if (gejala_dipilih[j] == True):
            jml_gejala_kasus = 0
            jml_gejala_dipilih = jml_gejala_dipilih +1
            for k in range(len(basis_kasus)):
                if (basis_kasus[k][0] == kasus[i]):
                    penyakit_kasus[i] = basis_kasus[k][1]
                    jml_gejala_kasus = jml_gejala_kasus + 1
                    if (basis_kasus[k][2] == gejala[j]):
                        jml_gejala_cocok = jml_gejala_cocok + 1
    print("=====", kasus[i], "====")
    print("Penyakit:", penyakit_kasus[i])
    print("Jml Gejala Dipilih:", jml_gejala_dipilih)
    print("Jml Gejala di Kasus:", jml_gejala_kasus)
    print("Jml Yg Cocok:",jml_gejala_cocok)
    pembagi = max(jml_gejala_dipilih,jml_gejala_kasus)
    print("Pembagi: MAX(",jml_gejala_dipilih,",",jml_gejala_kasus,")=",pembagi)
    if (pembagi == 0):
        hasil = 0
    else:
        hasil = jml_gejala_cocok / pembagi
    print("Hasil:",jml_gejala_cocok,"/",pembagi,"=",hasil)
    nilai_hasil.append(hasil)
print("")
print("Kasus:",kasus)
print("Hasil:",nilai_hasil)
kasusrangking = []
penyakitrangking = []
hasilrangking = []
for i in range(len(kasus)):
    kasusrangking.append(kasus[i])
    penyakitrangking.append(penyakit_kasus[i])
    hasilrangking.append(nilai_hasil[i])
for i in range(len(kasus)):
    for j in range(len(kasus)):
        if j > i:
            if hasilrangking[j]>hasilrangking[i]:
                tmphasil = hasilrangking[i]
                tmpkasus = kasusrangking[i]
                tmppenyakit = penyakitrangking[i]
                hasilrangking[i] = hasilrangking[j]
                kasusrangking[i] = kasusrangking[j]
                penyakitrangking[i] = penyakitrangking[j]
                hasilrangking[j] = tmphasil
                kasusrangking[j] = tmpkasus
                penyakitrangking[j] = tmppenyakit
print("Rangking Kasus:",kasusrangking)
print("Rangking Penyakit:",penyakitrangking)
print("Rangking Hasil:",hasilrangking)
```
